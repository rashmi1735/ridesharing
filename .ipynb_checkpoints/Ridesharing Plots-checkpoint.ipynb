```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
distanceSavedPercent_5 = 1.6806706024582798
distanceSavedPercent_10 = 30
distPlot = [distanceSavedPercent_5, distanceSavedPercent_10]
```

```python
tripsSavedPercent_5 = 47
tripsSavedPercent_10 = 30
tripPlot = [tripsSavedPercent_5, tripsSavedPercent_10]
```

```python
computationTime_5 = 2318.4961800575256
computationTime_10 = 4500
computationTimePlot = [computationTime_5 / 60, computationTime_10 / 60]
```

```python
xAxis = ('5 min', '10 min')
index = np.arange(len(xAxis))
```

```python
# distance saved plots
plt.bar(index, distPlot, align='center', alpha = 0.7)
plt.xticks(index, xAxis)
plt.ylabel('Distance Saved (%)')
plt.xlabel('Pool Times')
plt.title('Average distance saved per pool as a % of the total distance of individual trips,')

plt.show()
```

```python
# trips saved plot
plt.bar(index, tripPlot, align='center', alpha = 0.7)
plt.xticks(index, xAxis)
plt.ylabel('Trips Saved (%)')
plt.xlabel('Pool Times')
plt.title('Average number of trips saved per pool as a % of the number of individual trips')

plt.show()
```

```python
#computation time
plt.bar(index, computationTimePlot, align='center', alpha = 0.7)
plt.xticks(index, xAxis)
plt.ylabel('Computation Time (min)')
plt.xlabel('Pool Times')
plt.title('Computation Time per pool')

plt.show()
```
